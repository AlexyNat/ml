**Let's create a new feature.**

```python
def age_category(age):
    '''
    < 30 -> 1
    >= 30, <55 -> 2
    >= 55 -> 3
    '''
    if age < 30:
        return 1
    elif age < 55:
        return 2
    elif age >= 55:
        return 3
```

```python
age_categories = [age_category(age) for age in data.Age]
data['Age_category'] = age_categories
```

**Another way is to do it with `apply`.**

```python
data['Age_category'] = data['Age'].apply(age_category)
```

**5. Is that true that passengers younger than 30 y.o. survived more frequently than those older than 60 y.o.? What are shares of survived people among young and old people?**

- 40.6% among young and 22.7% among old

```python
young = (data[(data.Age < 30) & (data.Survived == 1)].shape[0] / data[(data.Age < 30)].shape[0]) * 100
old = (data[(data.Age > 60) & (data.Survived == 1)].shape[0] / data[(data.Age > 60)].shape[0]) * 100
print('{:.3}% among young and {:.3}% among old'.format(young, old))
```

**8. How is average age for men/women dependent on `Pclass`? Choose all correct statements:**
- On average, men of 1 class are older than 40
- On average, women of 1 class are older than 40
- Men of all classes are on average older than women of the same class
- On average, passengers of the first class are older than those of the 2nd class who are older than passengers of the 3rd class

```python
aver_age = data.groupby(['Pclass', 'Sex'])['Age'].mean()
display(aver_age)
print('On average, men of 1 class are older than 40: {}'.format(aver_age[1, 'male'] > 40))
print('On average, women of 1 class are older than 40: {}'.format(aver_age[1, 'female'] > 40))
print('Men of all classes are on average older than\
women of the same class: {}'.format((aver_age[:, 'male'] > aver_age[:, 'female']).all()))

# aver_age[1] > aver_age[2]).all() and (aver_age[2] > aver_age[3]).all()) by Male and Female
print('On average, passengers of the first class are older\
than those of the 2nd class who are older than passengers of the 3rd class: {}'.format(
    aver_age[1].max() > aver_age[2].max() > aver_age[3].max()))
```
